```python
import requests
import time
from bs4 import BeautifulSoup
from collections import defaultdict

# Initialize a dictionary to store article counts by year
year_count = defaultdict(int)

# Base URL for the site, will use page number to iterate
base_url = "https://www.aisixiang.com/zhuanti/263.html?page="

# Function to extract article years from a single page
def get_articles_years(page_number):
    url = f"{base_url}{page_number}"
    response = requests.get(url)

    if response.status_code != 200:
        print(f"Failed to retrieve page {page_number}")
        return []

    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all article elements, assuming they are within <li> elements
    articles = soup.find_all('li') #, class_='list-group-item'

    # Extract the year for each article
    years = []
    for article in articles:
        # Look for the publication year, assuming it is in a specific part of the article
        # In this case, we will assume the year is embedded in a <span> or <time> tag
        year_tag = article.find('span')  # Modify this based on actual HTML structure
        if year_tag:
            #print(year_tag)
            year = year_tag.text.strip().split('-')[0]  # Get the year part (e.g., "2024")
            #print(year)
            if year.isdigit():
                years.append(year)

    return years

# Loop through all 19 pages
for page in range(1, 20):
    print(f"Processing page {page}...")
    article_years = get_articles_years(page)
    time.sleep(2)

    # Update the year count dictionary
    for year in article_years:
        year_count[year] += 1

# Print the results
print("\nArticle count per year:")
for year, count in sorted(year_count.items()):
    print(f"{year}: {count} articles")
```

```python
import requests
from bs4 import BeautifulSoup
import re
from collections import defaultdict
from urllib.parse import urljoin

# Base URL for the column
base_url = "https://ciss.tsinghua.edu.cn/column/zmgx/"

# Dictionary to store year-wise frequency count
year_count = defaultdict(int)

# Function to extract year from article
def extract_year(article_html):
    # Assume that the year is mentioned somewhere in the article page, typically in a date format
    year_pattern = r"\b(20\d{2})\b"  # matches years from 2000 to 2099
    year_match = re.search(year_pattern, article_html)
    if year_match:
        return year_match.group(1)
    return None

# Function to check for the keyword in the article page
def check_for_keyword_in_article(article_url):
    response = requests.get(article_url)
    article_html = response.text
    # Check for the keyword '全球南方'
    if "全球南方" in article_html:
        year = extract_year(article_html)
        if year:
            year_count[year] += 1

# Iterate through all 15 pages (pages 1 to 15)
for page_num in range(1, 16):
    page_url = f"{base_url}{page_num}"
    response = requests.get(page_url)
    page_html = response.text
    soup = BeautifulSoup(page_html, 'html.parser')

    # Find all article links (these might be in <a> tags, so we need to find the right class or attribute)
    articles = soup.find_all('a', href=True)

    for article in articles:
        article_url = article['href']
        # Skip non-article links (e.g., anchors, javascript)
        if article_url.startswith('#') or article_url.startswith('javascript'):
            continue
        # Convert relative URLs to absolute URLs
        article_url = urljoin(base_url, article_url)

        # Check if the article URL is valid (e.g., it should point to a full page, not just a fragment)
        if article_url.startswith('https://ciss.tsinghua.edu.cn'):
            check_for_keyword_in_article(article_url)


# Print out the frequency of articles mentioning "全球南方" by year
for year, count in sorted(year_count.items()):
    print(f"Year: {year}, Count: {count}")
```
